```python
from alchemyapi import AlchemyAPI

import urllib.request
import urllib.parse
import csv, json
```

```python
api_key = ''
with open('api_key.txt', 'r') as apifile:
    api_key = apifile.readline().strip()
```

```python
url = "https://gateway-a.watsonplatform.net/calls/data/GetNews"
```

```python
params_object = {}
params_object['apikey'] = api_key
params_object['outputMode'] = 'json'
params_object['start'] = 'now-1d'
params_object['end'] = 'now'
params_object['q.enriched.url.enrichedTitle.relations.relation'] = '|action.verb.text=acquire,object.entities.entity.type=Company|'
params_object['return'] = 'enriched.url.title'

params = urllib.parse.urlencode(params_object)
req_url = url + "?%s" % params
with urllib.request.urlopen(req_url) as f:
    print(f.read().decode('utf-8'))
```
